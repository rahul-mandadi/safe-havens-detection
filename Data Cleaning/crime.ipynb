```python
crime_df.drop('1', axis=1, inplace = True)
```

```python
columns = crime_df.columns[0].split(',')
columns
```

```python
s = crime_df.iloc[0, 0]
```

```python
# Find the index of the first comma
first_comma_index = s.find(',')

# Find the index of the second comma
second_comma_index = s.find(',', first_comma_index + 1)
```

```python
s1 = s[:second_comma_index].split(',')
result = ','.join(s1)
s2 = s[second_comma_index:].split(',')
s2[0] = result
```

```python
col_name = crime_df.columns[0]
```

```python
def splitCols(row): 
    s = row[0]
    first_comma_index = s.find(',')
    second_comma_index = s.find(',', first_comma_index + 1)
    
    s1 = s[:second_comma_index].split(',')
    county_state = ','.join(s1).strip()
    
    s2 = s[second_comma_index + 1:].split(',')
    result = [county_state] + [value.strip() for value in s2] 
    return pd.Series(result)
```

```python
df1 = pd.DataFrame(columns=columns)
df1.head()
```

```python
df1 = crime_df.apply(splitCols, axis=1)
df1.head()
```

```python
df1.columns = columns
```
